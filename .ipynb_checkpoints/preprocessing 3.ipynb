```python
from tmdbv3api import Movie
tmdb_movie = Movie()
def get_genre(x):
    genres = []
    result = tmdb_movie.search(x)
    
    if not result:
        return np.NaN
    
    movie_id = result[0].id
    response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}'.format(movie_id, tmdb.api_key))
    data_json = response.json()
    
    if data_json['genres']:
        genre_str = " " 
        for i in range(0, len(data_json['genres'])):
            genres.append(data_json['genres'][i]['name'])
        return genre_str.join(genres)
    else:
        return np.NaN

```

```python
def get_director(x):
    if " (director)" in x:
        return x.split(" (director)")[0]
    elif " (directors)" in x:
        return x.split(" (directors)")[0]
    else:
        return x.split(" (director/screenplay)")[0]
```

```python


# Assuming you have NaN values represented as np.nan
df_2023['director_name'] = df_2023['Cast and crew'].map(lambda x: get_director(x) if isinstance(x, str) else np.nan)
```

```python
def get_actor1(x):
    return ((x.split("screenplay); ")[-1]).split(", ")[0])
```

```python
df_2023['actor_1_name'] = df_2023['Cast and crew'].map(lambda x: get_actor1(x) if isinstance(x,str) else np.nan)
```

```python
def get_actor2(x):
    if len((x.split("screenplay); ")[-1]).split(", ")) < 2:
        return np.NaN
    else:
        return ((x.split("screenplay); ")[-1]).split(", ")[1])
```

```python
df_2023['actor_2_name'] = df_2023['Cast and crew'].map(lambda x: get_actor2(x) if isinstance(x,str)else np.nan)
```

```python
def get_actor3(x):
    if len((x.split("screenplay); ")[-1]).split(", ")) < 3:
        return np.NaN
    else:
        return ((x.split("screenplay); ")[-1]).split(", ")[2])
```

```python
df_2023['actor_3_name'] = df_2023['Cast and crew'].map(lambda x: get_actor3(x)if isinstance(x,str)else np.nan)
```

```python
import numpy as np
from tmdbv3api import Movie
tmdb_movie = Movie()
import requests
import numpy as np

# Assuming df is your DataFrame with a 'Title' column
# and tmdb is your TMDb API instance
tmdb = TMDb()
tmdb_movie = Movie()



# Define the function to get the primary genre
def get_genres(x):
    try:
        result = tmdb_movie.search(x)

        if not result:
            return np.nan  # Handle the case where no result is found

        movie_id = result[0].id
        response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}'.format(movie_id, tmdb.api_key))
        data_json = response.json()

        if 'genres' in data_json and isinstance(data_json['genres'], list) and data_json['genres']:
            primary_genre = data_json['genres'][0]['name']  # Take the first genre as the primary genre
            return primary_genre
        
        else:
            return np.nan
    except Exception as e:
        print(f"Error processing {x}: {str(e)}")
        return np.nan

# Convert the 'Title' column to string, handle special characters
df['Title'] = df['Title'].astype(str)

# Add the primary genre directly to the 'genres' column
df['genres'] = df['Title'].apply(lambda x: get_genres(x))

# Display the DataFrame with the updated 'genres' column
print(df)









```
