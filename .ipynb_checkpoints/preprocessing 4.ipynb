```python
import pandas as pd
import numpy as np
import requests
import bs4 as bs
import urllib.request
```

## Extracting features of 2020 movies from Wikipedia

```python
link = "https://en.wikipedia.org/wiki/List_of_American_films_of_2021"
```

```python
source = urllib.request.urlopen(link).read()
soup = bs.BeautifulSoup(source,'lxml')
```

```python
tables = soup.find_all('table',class_='wikitable sortable')
```

```python
df1 = pd.read_html(str(tables[0]))[0]
df2 = pd.read_html(str(tables[1]))[0]
df3 = pd.read_html(str(tables[2]))[0]
df4 = pd.read_html(str(tables[3]).replace("'1\"\'",'"1"'))[0] # avoided "ValueError: invalid literal for int() with base 10: '1"'
```

```python
df = df1.append(df2.append(df3.append(df4,ignore_index=True),ignore_index=True),ignore_index=True)
```

```python
df_2021 = df[['Title','Cast and crew']]
```

```python
from tmdbv3api import TMDb
import json
import requests
tmdb = TMDb()
tmdb.api_key = '986edbcd3c49c744037b17ade99970a6'
```

```python
from tmdbv3api import Movie
tmdb_movie = Movie()
def get_genres(x):
    try:
        result = tmdb_movie.search(x)

        if not result:
            return np.nan  # Handle the case where no result is found

        movie_id = result[0].id
        response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}'.format(movie_id, tmdb.api_key))
        data_json = response.json()

        if 'genres' in data_json and isinstance(data_json['genres'], list) and data_json['genres']:
            primary_genre = data_json['genres'][0]['name']  # Take the first genre as the primary genre
            return primary_genre
        
        else:
            return np.nan
    except Exception as e:
        print(f"Error processing {x}: {str(e)}")
        return np.nan

# Convert the 'Title' column to string, handle special characters
df['Title'] = df['Title'].astype(str)
```

```python
df_2021['genres'] = df_2021['Title'].map(lambda x: get_genres(str(x)))
```

```python
df_2021
```

```python
def get_director(x):
    if " (director)" in x:
        return x.split(" (director)")[0]
    elif " (directors)" in x:
        return x.split(" (directors)")[0]
    else:
        return x.split(" (director/screenplay)")[0]
```

```python
df_2021['director_name'] = df_2021['Cast and crew'].map(lambda x: get_director(str(x)))
```

```python
def get_actor1(x):
    return ((x.split("screenplay); ")[-1]).split(", ")[0])
```

```python
df_2021['actor_1_name'] = df_2021['Cast and crew'].map(lambda x: get_actor1(str(x)))
```

```python
def get_actor2(x):
    if len((x.split("screenplay); ")[-1]).split(", ")) < 2:
        return np.NaN
    else:
        return ((x.split("screenplay); ")[-1]).split(", ")[1])
```

```python
df_2021['actor_2_name'] = df_2021['Cast and crew'].map(lambda x: get_actor2(str(x)))
```

```python
def get_actor3(x):
    if len((x.split("screenplay); ")[-1]).split(", ")) < 3:
        return np.NaN
    else:
        return ((x.split("screenplay); ")[-1]).split(", ")[2])
```

```python

df_2021['actor_3_name'] = df_2021['Cast and crew'].map(lambda x: get_actor3(str(x)))
```

```python
df_2021 = df_2021.rename(columns={'Title':'movie_title'})
```

```python
new_df21 = df_2021.loc[:,['director_name','actor_1_name','actor_2_name','actor_3_name','genres','movie_title']]
```

```python
new_df21 = new_df21.dropna(how='any')
```
